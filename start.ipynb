# Creating a Tensor

```python
# empty
a = torch.empty(2,3)
```

```python
type(a)
```

```python
# using zeroes 
# for initializing biases with 0 in NN
torch.zeros(2,3)
```

```python
# using ones
torch.ones(2,3)
```

```python
# using rand  0-1 value which will be random
torch.rand(2,3)
```

```python
# manual_seed
torch.manual(100)
torch.rand(2,3)
```

```python
# tensor
torch.tensor([[1,2,3], [3,4,5]])
```

```python
# arange
torch.arange(1,10,2)
```

```python
# linspace -> linearly space
torch.linspace(0,10,10)
```
