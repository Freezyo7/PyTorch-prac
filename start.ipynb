```python
# addition
# x+2
# substraction
x - 2
# multiplication
x * 3
# division
x / 4
# int division
(x*100) // 3
# mod
((x*100) // 3)%2
# power
x ** 2
```

# 3.) Reduction Operation

```python
e = torch.randint(size = (2,3), low = 0, high = 10)
f = torch.randint(size = (2,3), low = 0, high = 10, dtype = torch.float32)
```

```python
# sum
torch.sum(e)

# sum along column
torch.sum(e, dim = 0)

# sum along row
torch.sum(e, dim = 1)
```

```python
# mean
torch.mean(f)

# mean along column
torch.mean(f, dim = 0)
```

```python
# median
torch.median(f)

# median along column
torch.median(f, )
```

```python
# max and min
torch.max(f)

torch.min(f)
```

```python
# standard deviation
torch.std(f)
```

```python
# variance 
torch.var(f)
```

```python
# argmax - tell the position of largest element
torch.argmax(e)
```

```python
torch.argmin(e)
```

# 4.) Multiplication Operations 

```python
g = torch.randint(size = (2,3), low = 0, high = 15)
h = torch.randint(size = (3,2), low = 0, high = 15)

print(f"g: {g}\n --------------------------------------")
print(f"h: {h}")
```

```python
# multiplication
torch.matmul(g,h)
```

```python
vector1 = torch.tensor([1,2])
vector2 = torch.tensor([3,4])
```

```python
# dot product - a.b = [a1.b1+ a2.b2+ a3.b3]
torch.dot(vector1, vector2)
```

```python
# transpose
torch.transpose(g, 0, 1)
```

```python
i = torch.randint(size = (3,3), low = 0, high = 15, dtype=torch.float32)
i
```

```python
# determinant - square matrix det(A) = ad - bc
torch.det(i)
```

```python
# inverse
torch.inverse(i)
```
